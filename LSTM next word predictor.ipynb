```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer=Tokenizer()
```

```python
tokenizer.fit_on_texts([data])
```

```python
tokenizer.word_index
```

```python
len(tokenizer.word_index)
```

```python
for sentence in data.split('.'):
    print(sentence)
```

```python
for sentence in data.split('.'):
    print(tokenizer.texts_to_sequences([sentence])[0])
```

```python
input_sequences = []
for sentence in data.split('.'):
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1,len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

```python
max_len = max([len(x) for x in input_sequences])
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences=pad_sequences(input_sequences,maxlen=max_len,padding='pre')
```

```python
padded_input_sequences
```

```python
X=padded_input_sequences[:,:-1]
```

```python
y=padded_input_sequences[:,-1]
```

```python
X.shape
```

```python
y.shape
```

```python
from tensorflow.keras.utils import to_categorical
y=to_categorical(y,num_classes=326)
```

```python
y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
model=Sequential([
    Embedding(326, 50, input_length=max_len-1),
    LSTM(100,return_sequences=True),
    LSTM(100),
    Dense(326, activation='softmax')
    ])
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
model.build(input_shape=(None, max_len-1))
model.summary()
```

```python
model.fit(X,y,epochs=10)
```

```python
import time
import numpy as np

text = "The thieves arrived"

for i in range(10):
    token_text = tokenizer.texts_to_sequences([text])[0]
    padded_token_text = pad_sequences([token_text], maxlen=max_len-1, padding='pre')

    # Predict probabilities for next word
    preds = model.predict(padded_token_text, verbose=0)[0]

    # Sample instead of argmax
    next_index = np.random.choice(len(preds), p=preds)
    next_word = ''

    for word, index in tokenizer.word_index.items():
        if index == next_index:
            next_word = word
            break

    text += ' ' + next_word
    print(text)
    time.sleep(1)
```
