```python
import tensorflow as tf
from tensorflow import keras
from keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img=image.load_img(r'E:\Python\Basics of Python\Deep learning\data augumentation\train\poster.jpg',target_size=(200,200))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
```

```python
datagen=ImageDataGenerator(            #what transformations to apply
    rotation_range=30,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    width_shift_range=0.2,
    height_shift_range=0.2,
    fill_mode='constant'
)
```

```python
img=image.img_to_array(img)
```

```python
img.shape
```

```python
input_batch=img.reshape(1,200,200,3)
```

```python
i=0
for output in datagen.flow(input_batch,batch_size=1,save_to_dir='aug'):
    i=i+1

    if i==10:
        break
```
